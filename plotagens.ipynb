```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('analise1.csv')
df
```

```python
df = df[df['Algorithm'] != 'Branch-and-Bound']
twice = df[df['Algorithm'] == 'Twice around the tree']
chri = df[df['Algorithm'] == 'Christofides']

twice_time_euclidian = twice[twice['Dist_Type'] == 'Euclidean']['Time'].astype(np.float32).to_numpy()
chri_time_euclidian = chri[chri['Dist_Type'] == 'Euclidean']['Time'].astype(np.float32).to_numpy()

plt.title('Time comparison')
plt.xlabel('Number of vertices')
plt.ylabel('Time in seconds')

plt.scatter([2**i for i in range(4, 11)], twice_time_euclidian, label='Twice around the tree O(m*log(n))')
plt.scatter([2**i for i in range(4, 11)], chri_time_euclidian, label='Christofides O(n^3)')
plt.legend()

plt.show()
```

```python
twice = df[df['Algorithm'] == 'Twice around the tree']
chri = df[df['Algorithm'] == 'Christofides']

twice_time_euclidian = twice[twice['Dist_Type'] == 'Manhattan']['Time'].astype(np.float32).to_numpy()
chri_time_euclidian = chri[chri['Dist_Type'] == 'Manhattan']['Time'].astype(np.float32).to_numpy()

plt.title('Time comparison')
plt.xlabel('Number of vertices')
plt.ylabel('Time in seconds')

plt.scatter([2**i for i in range(4, 11)], twice_time_euclidian, label='Twice around the tree O(m*log(n))')
plt.scatter([2**i for i in range(4, 11)], chri_time_euclidian, label='Christofides O(n^3)')
plt.legend()

plt.show()
```

```python
df = pd.read_csv('analise2.csv')
df
```

```python
def series_to_numpy_time(df, algorithm):
    return df[df['Algorithm'] == algorithm]['Time'].astype(np.float32).to_numpy()

with open('metrics.csv', 'w') as f:
    f.write('Algorithm,Mean,Std,Dist_Type\n')
    
    df_euclid = df.loc[df['Dist_Type'] == 'Euclidean']

    bnb_time = series_to_numpy_time(df_euclid, 'Branch-and-Bound')
    twice_time = series_to_numpy_time(df_euclid, 'Twice around the tree')
    christofides_time = series_to_numpy_time(df_euclid, 'Christofides')

    f.write('{},{},{},{}\n'.format('Branch-and-Bound', bnb_time.mean(), bnb_time.std(ddof=1), 'Euclidean'))
    f.write('{},{},{},{}\n'.format('Twice around the tree', twice_time.mean(), twice_time.std(ddof=1), 'Euclidean'))
    f.write('{},{},{},{}\n'.format('Christofides', christofides_time.mean(), christofides_time.std(ddof=1), 'Euclidean'))

    df_man = df.loc[df['Dist_Type'] == 'Manhattan']

    bnb_time = series_to_numpy_time(df_man, 'Branch-and-Bound')
    twice_time = series_to_numpy_time(df_man, 'Twice around the tree')
    christofides_time = series_to_numpy_time(df_man, 'Christofides')

    f.write('{},{},{},{}\n'.format('Branch-and-Bound', bnb_time.mean(), bnb_time.std(ddof=1), 'Manhattan'))
    f.write('{},{},{},{}\n'.format('Twice around the tree', twice_time.mean(), twice_time.std(ddof=1), 'Manhattan'))
    f.write('{},{},{},{}\n'.format('Christofides', christofides_time.mean(), christofides_time.std(ddof=1), 'Manhattan'))
```

```python
def series_to_numpy_cost(df, algorithm):
    return df[df['Algorithm'] == algorithm]['Cost'].astype(np.float32).to_numpy()

with open('metrics2.csv', 'w') as f:
    f.write('Algorithm,Mean,Std,Dist_Type\n')
    
    df_euclid = df.loc[df['Dist_Type'] == 'Euclidean']

    bnb_cost = series_to_numpy_cost(df_euclid, 'Branch-and-Bound')
    twice_cost = series_to_numpy_cost(df_euclid, 'Twice around the tree') - bnb_cost
    christofides_cost = series_to_numpy_cost(df_euclid, 'Christofides') - bnb_cost
    

    #f.write('{},{},{},{}\n'.format('Branch-and-Bound', bnb_time.mean(), bnb_time.std(ddof=1), 'Euclidean'))
    f.write('{},{},{},{}\n'.format('Twice around the tree', twice_cost.mean(), twice_cost.std(ddof=1), 'Euclidean'))
    f.write('{},{},{},{}\n'.format('Christofides', christofides_cost.mean(), christofides_cost.std(ddof=1), 'Euclidean'))

    
    df_man = df.loc[df['Dist_Type'] == 'Manhattan']

    bnb_cost = series_to_numpy_cost(df_man, 'Branch-and-Bound')
    twice_cost = series_to_numpy_cost(df_man, 'Twice around the tree') - bnb_cost
    christofides_cost = series_to_numpy_cost(df_man, 'Christofides') - bnb_cost

    #f.write('{},{},{},{}\n'.format('Branch-and-Bound', bnb_time.mean(), bnb_time.std(ddof=1), 'Manhattan'))
    f.write('{},{},{},{}\n'.format('Twice around the tree', twice_cost.mean(), twice_cost.std(ddof=1), 'Manhattan'))
    f.write('{},{},{},{}\n'.format('Christofides', christofides_cost.mean(), christofides_cost.std(ddof=1), 'Manhattan'))
```
